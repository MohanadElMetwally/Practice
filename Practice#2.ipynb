```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tips['size'].unique()
```

```python
sns.relplot(x= 'total_bill',y= 'tip', data= tips)
plt.show()
```

```python
sns.relplot(x= 'total_bill',y= 'tip', data= tips, hue= 'smoker')
plt.show()
```

```python
sns.relplot(x= 'total_bill',y= 'tip', data= tips, hue= 'smoker', style= 'time')
plt.show()
```

```python
sns.relplot(x= 'total_bill',y= 'tip', data= tips, hue= 'smoker', style= 'time', size= 'size')
plt.show()
```

```python
sns.relplot(x= 'total_bill',y= 'tip', data= tips, hue= 'smoker', style= 'size', size= 'time')
plt.show()
```

```python
sns.relplot(x= 'total_bill',y= 'tip', data= tips, hue= 'smoker', style= 'time',kind='line')
plt.show()
```

```python
fmri = sns.load_dataset('fmri')

fmri
```

```python
sns.relplot(x= 'timepoint', y= 'signal', data= fmri, label='signal', kind= 'line')
```

```python
fmri[fmri['timepoint'] == 5].mean(numeric_only=True)
```

```python
sns.relplot(x= 'timepoint', y= 'signal', data= fmri, label='signal', kind= 'line',hue= 'event',style='region',errorbar=('ci',False))
```

```python
dots = sns.load_dataset('dots').query('align == "dots"')
dots
```

```python
sns.relplot(x= 'time', y= 'firing_rate',data= dots,kind='line')
```

```python
sns.relplot(x= 'time', y= 'firing_rate',data= dots, kind='line', hue= 'coherence')
```

```python
palette = sns.cubehelix_palette(light= 0.8, n_colors=6)
sns.relplot(x= 'time', y= 'firing_rate',data= dots, kind='line', hue= 'coherence',style= 'choice', height=6, aspect= 2, palette=palette)
plt.grid()
```

```python
sns.relplot(x= 'total_bill', y= 'tip', data= tips, hue='smoker', col='time',height=6)
```

```python
sns.relplot(x= 'total_bill', y= 'tip', data= tips, hue='smoker', col='time',row='size',height=6)
```

```python
sns.relplot(x= 'total_bill', y= 'tip', data= tips, hue='smoker', col='size', col_wrap= 2)
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(x= 'timepoint', y= 'signal', data= fmri, style= 'event', hue= 'region', markers= True, errorbar=('ci', 68), err_style='bars')
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(x= 'timepoint', y= 'signal', data= fmri.query("region == 'frontal'"), hue= 'event', markers= True, errorbar=('ci', 68), err_style='bars')
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(x= 'timepoint', y= 'signal', data= fmri[fmri['region'] == 'frontal'], hue= 'event', markers= True, errorbar=('ci', 68), err_style='bars')
```
