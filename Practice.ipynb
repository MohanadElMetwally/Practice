```python
movies = pd.read_csv('IMDB-Movie-Data.csv', nrows=50)
movies
```

```python
plt.figure(figsize=(20,8))
plt.plot(movies.Rank, movies['Revenue (Millions)'])
```

```python
idx = movies['Revenue (Millions)'].idxmax()
movies.iloc[idx] # type: ignore
```

```python
plt.figure(figsize=(15,8))
plt.scatter(movies.Rank, movies.Rating) # type: ignore
```

```python
plt.figure(figsize=(15,8))
plt.bar(movies.Rank, movies.Rating)
plt.show()
```

```python
plt.figure(figsize=(20,10))
#First subplot Rank vs Rating
plt.subplot(2,2,1)
plt.bar(movies.Rank, movies.Rating)
plt.xlabel('Rank')
plt.ylabel('Rating')
plt.title('Rank vs Rating')
#Second subplot Rank vs Revenue
plt.subplot(2,2,2)
plt.scatter(movies.Rank, movies['Revenue (Millions)'])
plt.xlabel('Rank')
plt.ylabel('Revenue')
plt.title('Rank vs Revenue')
#Third subplot Runtime vs Votes
plt.subplot(2,2,3)
plt.bar(movies['Runtime (Minutes)'], movies.Votes, width= 2)
plt.xlabel('Runtime')
plt.ylabel('Votes')
plt.title('Runtime vs Votes')
#Fourth subplot Votes vs Revenue 
plt.subplot(2,2,4)
plt.bar(movies['Revenue (Millions)'], movies['Votes'], width=10)
plt.xlabel('Revenue')
plt.ylabel('Votes')
plt.title('Votes vs Revenue')

plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(20,10))

ax[0][0].bar(movies.Rank, movies.Rating)
ax[0][0].set_xlabel('Rank')
ax[0][0].set_ylabel('Rating')
ax[0][0].set_title('Rank vs Rating')

ax[0][1].scatter(movies.Rank, movies['Revenue (Millions)'])
ax[0][1].set_xlabel('Rankax')
ax[0][1].set_ylabel('Revenueax')
ax[0][1].set_title('Rank vs Revenueax')

ax[1][0].bar(movies['Runtime (Minutes)'], movies.Votes, width= 2)
ax[1][0].set_xlabel('Runtime')
ax[1][0].set_ylabel('Votes')
ax[1][0].set_title('Runtime vs Votes')

ax[1][1].bar(movies['Revenue (Millions)'], movies['Votes'], width=10)
ax[1][1].set_xlabel('Revenue')
ax[1][1].set_ylabel('Votes')
ax[1][1].set_title('Votes vs Revenue')

plt.show()
```
